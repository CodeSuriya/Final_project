2. Create a figure, Plot Delivery and Restaurant Locations, and Customize Plot

```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(10, 6))

# Plot delivery locations (blue circles)
plt.scatter(delivery_lon, delivery_lat, c='blue', s=80, label='Delivery Locations', alpha=0.6)

# Plot restaurant location (red pentagon)
plt.scatter(restaurant_lon, restaurant_lat, c='red', marker='p', s=150, label='Restaurant')

# Customize the plot
plt.title('Bangalore Delivery Locations and the Restaurant', fontsize=14)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(loc='best')
plt.grid(True)

# Show the plot
plt.show()
```

3. Set Geographical Limits and Save Figure


```python

# Create figure
plt.figure(figsize=(10, 6))

# Plot delivery locations
plt.scatter(delivery_lon, delivery_lat, c='blue', s=80, label='Delivery Locations', alpha=0.6)

# Plot restaurant location
plt.scatter(restaurant_lon, restaurant_lat, c='red', marker='p', s=150, label='Restaurant')

# Set geographical limits
plt.xlim(min(delivery_lon + [restaurant_lon]) - 0.01, max(delivery_lon + [restaurant_lon]) + 0.01)
plt.ylim(min(delivery_lat + [restaurant_lat]) - 0.01, max(delivery_lat + [restaurant_lat]) + 0.01)

# Customize the plot
plt.title('Bangalore Delivery Locations and the Restaurant', fontsize=14)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(loc='best')
plt.grid(True)

# Save the figure as an image
plt.savefig('Bangalore.png', dpi=300)

# Show the plot
plt.show()
```
